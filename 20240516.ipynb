````python
import requests
from bs4 import BeautifulSoup
url = "https://www.irasutoya.com/2013/10/blog-post_9165.html"
r = requests.get(url)
sp = BeautifulSoup(r.text, 'html.parser')
sp.title.text
````

````python
for img in sp.find('div', class_='separator').find_all('img'):
    # print(img)
    print(img.get('src'))
````

````python
# 任務一：請將動物圖片儲存在<img01>資料夾
# 檔名格式：img[流水號].png
i = 1
for img in sp.find('div', class_='separator').find_all('img'):
    imgurl = img.get('src')
    r = requests.get(imgurl)
    with open(f'img01/img{i:02}.png', 'wb') as f:
        f.write(r.content)
    i = i + 1
print('任務結束')
````

````python
# 任務二：請將動物圖片儲存在<img02>資料夾
# 檔名格式：原始檔名
for img in sp.find('div', class_='separator').find_all('img'):
    imgurl = img.get('src')
    filename = img.get('src').split('/')[-1]
    r = requests.get(imgurl)
    with open(f'img02/{filename}', 'wb') as f:
        f.write(r.content)
print('任務結束')
````

````python
# 顯示多張圖片
import matplotlib.pyplot as plt
import matplotlib.image as img
import os
def imgridshow(imgurl, c=5):
    plt.figure(figsize=(16,9))
    images = os.listdir(imgurl)
    t = len(images)
    r = t // c
    if t % c != 0:
        r += 1
    for i in range(t):
        plt.subplot(r,c,i+1)
        plt.axis('off')
        plt.imshow(img.imread(imgurl + "/" + images[i]))
    plt.show()
````

## WebAPP 的開發

````python
def catOrDog(a):
    if 'cat' == a:
        url = "https://api.thecatapi.com/v1/images/search"
        r = requests.get(url)
        datas = r.json()
        imgurl = datas[0]['url']
    else:
        url = "https://dog.ceo/api/breeds/image/random"
        r = requests.get(url)
        datas = r.json()
        imgurl = datas['message']
    return f"<img src={imgurl}>"

HTML(catOrDog('cat'))
````
